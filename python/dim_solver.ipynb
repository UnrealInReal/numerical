Set $f(x) = 54 x^6 + 45 x^5 − 102 x^4 − 69 x^3 + 35 x^2 + 16 x − 4$. 

Plot the function on the interval [−2,2], and use the Secant Method to find all five roots in the interval.

To which of the roots is the convergence linear, and to which is it superlinear?

```python
from math import pow as Pow
def y(x):
    return 54 * Pow(x, 6) + 45 * Pow(x, 5) - 102 * Pow(x, 4) - 69 * Pow(x, 3) + 35 * Pow(x, 2) + 16 * x - 4
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(-2, 2, 0.001)
y = [y(a) for a in x]

plt.plot(x,y)
plt.ylim([-0.5,0.5])
plt.xlim([-2, 2])
plt.show()
```
